# Исследование объявлений о продаже квартир

В нашем распоряжении архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Изучение данных из файла

```python
import pandas as pd
import numpy as np

# visualization
import matplotlib.pyplot as plt 
import seaborn as sns
#Q-Q-plot
import statsmodels.api as sm

# correlation matrix
import phik
from phik.report import plot_correlation_matrix
from phik import report

import warnings
warnings.filterwarnings("ignore")
```

```python
STATE = 2222
```

```python
def first_glance(directory, title, **kwargs):
    '''
    Загружает данные; 
    Печатает: 
            Общая информация, 
            5 случайных строк, 
            Количество дубликатов, 
            Количество и процентное соотношение пропусков.
    
    Параметры
    ----------
    directory: str
        Расположение csv файла
    title: str
        Название файла
        
    Возвращает 
    ----------
    DataFrame
    
    '''
    data = pd.read_csv(directory, **kwargs)
    display(title)
    print('-------------Общая информация---------------')
    display(data.info())

    print('------------5 случайных строк---------------') 
    display(data.sample(5, random_state=STATE))
    print()
    
    if data.duplicated().sum().sum() != 0:
        display('-------Количество дубликатов---------', 
        data.duplicated().sum())
    else:
        print('Полные дубликаты отсутствуют')

    if data.isna().sum().sum() != 0:
        # создает датафрейм с количеством пропусков в каждой колонке
        na_values = data.isna().sum().to_frame().rename(columns = {0: 'пропуски'})
        # добавляет новую колонку с процентом пропусков
        na_values['% от общего количества'] = (na_values['пропуски'] / data.shape[0]).round(2)
        # сортирует таблицу по убыванию количества пропусков
        display('--Количество и процентное соотношение пропусков--',
                na_values.sort_values(by = 'пропуски', ascending = False))
    else:
        print('Пропуски отсутствуют')

    print()
    
    print('----------Описательная статистика----------')
    display(data.describe())
    
    return data
```

```python
df = first_glance('real_estate_data.csv', 'Real_estate', sep='\t')
```

```python
# группировка колонок по двум типам: числовые и категориальные
def group_columns(df):
    numeric_columns = []
    categorical_columns = []

    for column in df.columns:
        if df[column].dtype.name in ['int64', 'float64', 'bool']:
            numeric_columns.append(column)
        if df[column].dtype.name in ['object', 'category', 'datetime64']:
            categorical_columns.append(column)  
            
    return categorical_columns, numeric_columns
```

```python
cat_columns, num_columns = group_columns(df)
```

```python
def visualize_info(df, columns):
    '''
    В виде подграфиков:
        Выводит информацию describe;
        Строит графики:
            диаграмма размаха;
            гистограмма;
            график квантиль-квантиль.
    
     Параметры
     ----------
     df: DataFrame
         таблица для графика
     columns: str
         названия колонок с численными данными
    '''
    fig, ax = plt.subplots(len(columns), 4, gridspec_kw={'wspace': 0.5, 'hspace': 0.3})
    fig.set_figheight(5 * len(columns))
    fig.set_figwidth(17)
    
    for i, column in enumerate(columns):
        ax_plot = [ax[i,0], ax[i,1], ax[i,2], ax[i,3]]

        # describe info
        describe_series = round(df[column].describe())
        # создание таблицы на основе перевернутой series.describe() 
        describe_table = ax_plot[0].table(
            cellText=describe_series.values.reshape(describe_series.shape[0],1),
            rowLabels=list(describe_series.index),
            colLabels=[column],
            colWidths = [0.9],
            loc = 'center'
                                         )
        ax_plot[0].grid(False)
        ax_plot[0].axis('off')
        describe_table.scale(1, 2)
        describe_table.auto_set_font_size(False)
        describe_table.set_fontsize(15)
        
        # violin plot
        sns.violinplot(data=df[column], ax=ax_plot[1], color='salmon')
        ax_plot[1].set_xlabel(column, fontsize = 15)
        
        # density plot
        sns.distplot(a=df[column], kde=True, ax=ax_plot[2], color='olive')
        ax_plot[2].set_xlabel(column, fontsize = 15)
        ax_plot[2].set_ylabel('Density', fontsize = 14)
        
        # qq-plot
        sm.qqplot(df[column].dropna().astype(np.float64), ax=ax_plot[3], fit=True, line='45')
        ax_plot[3].set_xlabel('Theoretical Quantiles', fontsize = 15)
        ax_plot[3].set_ylabel('Sample Quantiles', fontsize = 15)
```

```python
df = visualize_info(df, num_columns)
```

Так как в данных отсутствует единый вид записи для всех населенных пунктах, изменим некоторые значения для точности анализа в дальнейшем.

```python
replace_names = ['село ', 'поселок ', 'посёлок ', 'деревня ', 'пок ', 'станции']
for name in replace_names:
    # заменяем  replace_names на пустую строку, чтобы отображалось только название населенного пункта
    df.loc[:, 'locality_name'] = df.loc[:, 'locality_name'].str.replace(name, '') 

replace_gardens = ['садоводческое некоммерческое товарищество ', 'садовое товарищество ']
for name in replace_gardens:
    # заменяем  replace_gardens на снт, чтобы оно отображалось перед названием населенного пункта
    df.loc[:, 'locality_name'] = df.loc[:, 'locality_name'].str.replace(name, 'снт ')

# заменяем 'коттеджный ' на 'коттеджный поселок, чтобы оно отображалось перед названием населенного пункта
df.loc[:, 'locality_name'] = df.loc[:, 'locality_name'].str.replace('коттеджный ', 'коттеджный поселок ')

df.groupby('locality_name')['locality_name'].count()
```

Из 365 населенных пунктов осталось 322, замена проведена успешно.

Заполним пропуски в столбцах с жилой площадью и площадью кухни. Для более точного рассчета будем исходить из указанной общей площади недвижимости.

```python
# округляем значения до десятков для применения функции к полученным данным, изменяем тип данных для экономии памяти
df['total_area_rounded'] = df['total_area'].round(decimals=-1).astype('Int16')
df.groupby('total_area_rounded')['total_area_rounded'].count()
```

```python
df['kitchen_area'].isna().sum()
```

```python
# функция для поиска соответствующих строк со схожей площадью (в диапазоне десяти квадратных метров) 
# и присвоение медианы площади кухни этих групп для заполнения пропусков
def kitchen_area(row): 
    area = row['total_area_rounded']
    count = df.loc[area == df['total_area_rounded'], 'kitchen_area'].count()
    median = df.loc[(area == df['total_area_rounded']) & (count != 0), 'kitchen_area'].median()

    return median

df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = df.loc[df['kitchen_area'].isna()].apply(kitchen_area, axis=1)
# проверка
df['kitchen_area'].isna().sum()
```

С учётом графика ниже, мы видим, что при общей площади свыше 400 размеры кухонной зоны остаются в одном диапазоне и практически не увеличиваются (за исключением выбросов), значит можно найти медиану площади кухни для всех помещений с площадью выше 400 квадратных метров и заполнить пропуски данной медианой. 

```python
df.plot(x='total_area' , y='kitchen_area', kind='scatter', alpha=0.2)
plt.xlabel('Общая площадь (м2)')
plt.ylabel('Площадь кухни (м2)')
plt.title('Зависимость площади кухни от общей площади')
plt.show()
```

Обрабатываем пропуски в жилой площади исходя из аналогичных объявлений.

```python
df['living_area'].isna().sum()
```

```python
def living_area(row): 
    area = row['total_area_rounded']
    count = df.loc[area == df['total_area_rounded'], 'living_area'].count()
    median = df.loc[(area == df['total_area_rounded']) & (count != 0), 'living_area'].median()

    return median

df.loc[df['living_area'].isna(), 'living_area'] = df.loc[df['living_area'].isna()].apply(living_area, axis=1)
df['living_area'].isna().sum()
```

Заполняем оставшиеся пропуски с помощью коэффициента корреляции общей и жилой площади.

```python
# находим коэффициент корреляции
corr_living_area = df['living_area'].corr(df['total_area'])
# заполняем пропуски в столбце с жилой площадью, перемножив общую площадь и коэффициент
df.loc[df['living_area'].isna(), 'living_area'] = df.loc[df['living_area'].isna(), 'total_area'] * corr_living_area
```

Пропуски в данных могли быть вызваны отсутствуем некоторых параметров изначально, при заполнении формы подачи объявления. Некоторые параметры возможно добавились позже или изначально рассчитываются автоматически, тогда отсутствие гео-данных об этих объектах или некорректно указанный адрес недвижимости могли привести к пропускам.

### Обработка нулевых значений

Избавимся от нулей в столбце с количеством комнат.

```python
# проверим, сколько объектов с нулями в столбце с количеством комнат
df.loc[df['rooms'] == 0]['rooms'].count()
```

```python
df.loc[df['rooms'] == 0]['rooms'].count()
```

Предположим, что объекты недвижимости типа "студия" содержат нули в столбце с количеством комнат и заменим их на единицы, так как формально вся квартира является комнатой.

```python
# проверка количества 
df.loc[(df['studio'] == True) & (df['rooms'] == 0)]['rooms'].count()
# замена на 1
df.loc[(df['studio'] == True) & (df['rooms'] == 0), 'rooms'] = 1
```

Посмотрим на оставшиеся объявления, не удовлетворяющие условие.

```python
df.loc[df['rooms'] == 0].groupby('total_area')['total_area'].count()
```

Заменим нули на единицы в тех объявлениях где общая площадь не больше 50, так как, вероятнее всего эти объекты содержат всего одну комнату.

```python
df.loc[(df['total_area'] < 50) & (df['rooms'] == 0), 'rooms'] = 1
df.loc[df['rooms'] == 0].groupby('total_area')['total_area'].count()
```

Заменим оставшееся нулевое значение медианой аналогичных по площади объектов недвижимости.

```python
df.loc[df['rooms'] == 0] = df.loc[(df['total_area'] > 300) & (df['total_area'] < 400)]['rooms'].median()
# проверка
df.loc[df['rooms'] == 0]
```

### Вывод

Мы заменили нулевые значения в столбце с количеством комнат. Возможно появление нулей в данной столбце было вызвано некорректностью внесения начальных данных.

Переведем строковое значение даты публикации объявления в тип даты и округлим её до суток для простоты отображения и группировок.

Изменим тип данных на целочисленный в столбцах с этажами, балконами, комнатами, сроком размещения, парками, водоёмами, аэропортами и расстоянием до ближайших объектов, так как вещественный тип данных в этих столбцах является избыточным и затрудняет работу с информацией.

```python
df.describe().T
```

```python
df['floor'] = df['floor'].astype('Int8')
df['floors_total'] = df['floors_total'].astype('Int8')
df['balcony'] = df['balcony'].astype('Int8')
df['rooms'] = df['rooms'].astype('Int8')

# округлям значения и изменяем тип данных
df['days_exposition'] = df['days_exposition'].round().astype('Int64')
df['cityCenters_nearest'] = df['cityCenters_nearest'].round().astype('Int64')
df['parks_around3000'] = df['parks_around3000'].round().astype('Int64')
df['parks_nearest'] = df['parks_nearest'].round().astype('Int64')
df['ponds_around3000'] = df['ponds_around3000'].round().astype('Int64')
df['ponds_nearest'] = df['ponds_nearest'].round().astype('Int64')
df['airports_nearest'] = df['airports_nearest'].round().astype('Int64')
```

## Расчёты и добавление результатов в таблицу

Рассчитаем расстояние до центра в километрах и добавим его в новый столбец.

```python
df.loc[:, 'cityCenters_km'] = df.loc[:,'cityCenters_nearest'] / 1000
df['cityCenters_km'] = df['cityCenters_km'].astype('Int16')
df.head()
```

Добавим в таблицу столбец с ценой за квадратный метр.

```python
df['sq_meter_price'] = df['last_price'] / df['total_area']
df['sq_meter_price'] = df['sq_meter_price'].round(decimals=2).astype('float32')
df.head()
```

Добавим в таблицу три новых столбца - день недели, месяц и год размещения объявления.

```python
# создаем новые столбцы, округляем значения и меняет их тип на более подходящий
df['day_of_the_week'] = df['first_day_exposition'].dt.weekday.astype('float16').round().astype('Int16')
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year

df['month'].astype('float16').round().astype('Int16')
df['year'].astype('float16').round().astype('Int16')
df.head()
```

Разделим указанные этажи на три группы - первый, последний и другой.

```python
df.loc[df['floor'] == 1, 'floor_group'] = 'первый'
df.loc[df['floor'] == df['floors_total'], 'floor_group'] = 'последний'
df.loc[(df['floor'] != df['floors_total']) & (df['floor'] != 1), 'floor_group'] = 'другой'
df.head()
```

Рассчитаем соотношение общей площади и жилой, а также общей площади и кухонной зоны. Добавим их в таблицу для дальнейшего анализа.


```python
# функция для расчета соотношения в конкретной строке и создания нового столбца с полученным соотношением
def corr_total_living(row):
    row['corr_total_living'] = row['living_area'] / row['total_area']
    return row['corr_total_living']

df.loc[:, 'corr_total_living'] = df.apply(corr_total_living, axis=1)
```

```python
# функция для расчета соотношения в конкретной строке и создания нового столбца с полученным соотношением
def corr_total_kitchen(row):
    row['corr_total_kitchen'] = row['kitchen_area'] / row['total_area']
    return row['corr_total_kitchen']

df.loc[:, 'corr_total_kitchen'] = df.apply(corr_total_kitchen, axis=1)
```

В таблицу были добавлены следующие столбцы:
- цена за квадратный метр
- день недели размещения
- месяц размещения
- год размещения
- группа по этажу квартиры (первый, последний, другой)
- соотношение жилой площади к общей
- соотношение площади кухни к общей
- расстояние до центра в километрах

Данная информация позволит найти новые зависимости и сделать более точные расчеты.

```python
# функция для построения диаграмм рассеивания
# параметр - столбец с которым сравниваем значения цены и датафрейм для построения сводной таблицы
# строим сводную таблицу по медианам, избавляемся от мультииндекса и строим график

def plot_scatter(data, column):
    price_rooms = data.pivot_table(index=column, values='last_price', aggfunc=['median']).reset_index()
    price_rooms.columns = [column, 'last_price']
    price_rooms.plot(x=column, y='last_price', kind='scatter', alpha=0.3, figsize=(10, 5))
```

```python
plot_scatter(df_clean, 'total_area')
plt.xlabel('Общая площадь (м2)')
plt.ylabel('Цена (10 млн)')
plt.title('Зависимость цены от общей площади')
plt.show()
```

График показывает, что цена довольно сильно зависит от общей площади, 
однако с увеличением площади появляются сильные отклонения в цене.

```python
# функция для построения столбчатых диаграмм
# параметры - столбец с которым сравниваем значения цены и датафрейм для построения сводной таблицы
# строим сводную таблицу по медианам, избавляемся от мультииндекса и строим график

def plot_bar(data, column):
    price_rooms = data.pivot_table(index=column, values='last_price', aggfunc=['median']).reset_index()
    price_rooms.columns = [column, 'last_price']
    price_rooms.plot(x=column, y='last_price', kind='bar', figsize=(8, 5), alpha=0.7)
```

```python
plot_bar(df_clean, 'rooms')
plt.xlabel('Количество комнат')
plt.ylabel('Цена (10 млн)')
plt.title('Зависимость цены от количества объявлений')
plt.show()
```

С ростом количества комнат увеличивается цена, но цена перестаёт расти столь заметно, начиная с семи комнат.

```python
plot_bar(df_clean, 'floor_group')
plt.xlabel('Этаж')
plt.ylabel('Цена (10 млн)')
plt.title('Зависимость цены от этажа')
plt.show()
```

Вполне ожидаемо, стоимость жилья на первых этажах является самой низкой, далее идут последние этажи, 
а самая дорогая группа - остальные этажи.

```python
plot_bar(df_clean, 'day_of_the_week')
plt.xlabel('День недели')
plt.ylabel('Цена (10 млн)')
plt.title('Зависимость цены от дня недели (0-понедельник)')
plt.show()
```

Цена несильно зависит от дня размещения объявления, немного выше цена у размещенных во вторник и среду объявлений, а самая низкая - в субботу.

```python
plot_bar(df_clean, 'month')
plt.xlabel('Месяц')
plt.ylabel('Цена')
plt.title('Зависимость цены от месяца (1-январь)')
plt.show()
```

Цена немного изменяется в течение года, самая высокая цена за недвижимость оказалась в апреле, а самой низкой в июне.

```python
plot_bar(df_clean, 'year')
plt.xlabel('Год')
plt.ylabel('Цена (10 млн)')
plt.title('Зависимость цены от года размещения объявления')
plt.show()
```
