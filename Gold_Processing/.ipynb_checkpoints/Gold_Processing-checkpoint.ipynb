```python
# функция для проверки колонок таблицы на процент пропусков
# на вход принимает таблицу с данными
def check_isna(dataframe):
    columns = list(dataframe.columns)
    count_isna = 0 
    for column in columns:
        count_isna += dataframe[column].isna().sum()
    if count_isna == 0:
        print('Пропуски отсутствуют.')
    else:
        print('Суммарное количество пропусков во всех колонках', count_isna)
```

### Расчёт эффективности обогащения

Проверим расчет эффективности обогащения на общей выборке. 

Для этого рассчитаем признак "rougher.output.recovery" самостоятельно и сравним эти значения с имеющимся признаком с помощью MAE (средняя абсолютная ошибка) по формуле ниже.

<div>
<img src="https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg" width="500" height="300"/>
</div>

```python
concentrate = df['rougher.output.concentrate_au'] # C
feed = df['rougher.input.feed_au'] # F
tail = df['rougher.output.tail_au'] # T
recovery_original = df['rougher.output.recovery'] 
check = concentrate, feed, tail, recovery_original
```

```python
# проверка пропусков в используемых столбцах
for item in check:
    print(item.isna().sum())
```

```python
recovery_check = (concentrate * (feed - tail)) / (feed * (concentrate - tail)) * 100
```

```python
mae = mean_absolute_error(recovery_original, recovery_check).round()
print('MAE:', mae)
```

### Вывод

Отсутствующие признаки имеют тип параметра "output" — параметр конечного продукта и

их нет в тестовой выборке, на которой проводится финальное тестирование модели.

Пропущенные значения были заполнены методом интерполяции, т.к. данные признаков фиксируются с течением времени.

Рассчет эффективности показал, что исходный признак "rougher.output.recovery" имеет высокую точность, т.к. метрика MAE (средняя абсолютная ошибка) исходного признака и рассчитанного очень мала.

```python
# функция для построения распределений по двум выборкам
def hist(df1, df2, name, scale):
    print(name)
    plt.figure(figsize=(10, 6))

    sns.histplot(df1,
                 color="darkkhaki", 
                 label="Обучающая выборка", 
                 kde=True
                ).set(
        title='Распределение размеров гранул сырья на обучающей и тестовой выборках', 
        xlabel='Размер гранул сырья (мкм)',
        ylabel='Частота')

    sns.histplot(df2, 
                 color="deepskyblue", 
                 label="Тестовая выборка", 
                 kde=True)
    plt.axis(scale)
    plt.legend()
    plt.show()
```

### Суммарная концентрация веществ на разных этапах обработки руды

Проверим суммарную концентрацию веществ на разных этапах обработки руды.

```python
rougher_input_feed = pd.Series(
                    df_train['rougher.input.feed_au'] + 
                    df_train['rougher.input.feed_ag'] + 
                    df_train['rougher.input.feed_pb']
                    )
rougher_output_concentrate = pd.Series(
                    df_train['rougher.output.concentrate_au'] + 
                    df_train['rougher.output.concentrate_ag'] + 
                    df_train['rougher.output.concentrate_pb'] +
                    df_train['rougher.output.concentrate_sol']
                    )

primary_cleaner_output_concentrate = pd.Series(
                    df_train['primary_cleaner.output.concentrate_au'] +
                    df_train['primary_cleaner.output.concentrate_ag'] +
                    df_train['primary_cleaner.output.concentrate_pb'] +
                    df_train['primary_cleaner.output.concentrate_sol']
                    )
final_output_concentrate = pd.Series(
                    df_train['final.output.concentrate_au'] +
                    df_train['final.output.concentrate_ag'] +
                    df_train['final.output.concentrate_pb'] +
                    df_train['final.output.concentrate_sol']
                    )
```

## Модель

### Расчёт sMAPE

В качестве основной метрики для обучения моделей возьмем 
sMAPE - симметричную среднюю абсолютную ошибку в процентах. 

Данную метрику рассчитаем по следующим формулам:

<div>
<img src="https://pictures.s3.yandex.net/resources/smape_1576239058.jpg" width="700" height="500"/>
</div>

<div>
<img src="https://pictures.s3.yandex.net/resources/_smape_1576238814.jpg" width="700" height="500"/>
</div>



```python
# подсчет метрики smape
# на вход принимает целевое значение и значение предсказания
# т.к. у нас два целевых признака, предсказания также состоят из двух признаков
def smape(y_target, y_predict):
    error = abs(y_target - y_predict)
    scaling = (abs(y_target) + abs(y_predict)) / 2
    smape = (error / scaling).mean() * 100
    
    # подсчет суммарного smape для двух признаков
    final_smape = 0.25 * smape[0] + 0.75 * smape[1]
    
    return final_smape
```

```python
# создаем счетчик подсчета метрики smape для модели 
scorer = make_scorer(smape, greater_is_better=False)
```

### Модель случайного леса

```python
# обучаем модели со случайными гиперпараметрами, 
# чтобы сравнивать с ней дальнейшие модели
model_forest = RandomForestRegressor(n_estimators=50, max_depth=3, random_state=1)
model_forest.fit(features_train, target_train)
final_smape_forest = abs(cross_val_score(model_forest, 
                              features_train, 
                              target_train, 
                              cv=5, 
                              scoring=scorer).mean()).round(2)
final_smape_forest
```

#### Подбор гиперпараметров

```python
# создаем словарь для поиска лучших гиперпараметров
random_grid_forest = {
               'max_depth': [1, 2, 3, 4, 5, 10],
               'min_samples_leaf': [1, 2, 4],
               'min_samples_split': [2, 5, 10],
               'n_estimators': [10, 20, 30, 40, 50, 70],
               'bootstrap': [True, False]}
# задаем параметры поиска со своим счетчиком метрики
search = RandomizedSearchCV(model_forest, 
                            random_grid_forest, 
                            n_iter=100, 
                            scoring=scorer, 
                            n_jobs=-1, 
                            cv=5, 
                            random_state=1)
# запускаем поиск
result_forest = search.fit(features_train, target_train)
best_params_forest = result_forest.best_params_

print('Лучший sMAPE случайного леса', abs(result_forest.best_score_).round(2))
print('Лучшие гиперпараметры', best_params_forest)
```
